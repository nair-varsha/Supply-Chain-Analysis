# **<font color="Green">**<font face="Times New Roman">9. Visualizing Number of Products respect to different states
## **</font></font>**

```python
##  USING SEABORN

# Font styles
font1={'family':'serif','color':'r','size':15}
font2={'family':'serif','color':'b','size':10}

# Grouping by state and summing the quantities
state_sales = df.groupby('LOCATION')['NUMBER_OF_PRODUCTS_SOLD'].sum().reset_index()

# Set style
sns.set(style="whitegrid")

# Create a barplot
plt.figure(figsize=(6, 6))
sns.barplot(x='LOCATION', y= 'NUMBER_OF_PRODUCTS_SOLD', data=state_sales, palette="viridis")

plt.title("TOTAL PRODUCTS SOLD PER STATE",fontdict=font1)
plt.xlabel('LOCATION',fontdict=font2)
plt.ylabel('NUMBER_OF_PRODUCTS_SOLD',fontdict=font2)

plt.show()
```

# **<font color="Green">**<font face="Times New Roman">10. Visualizing Transportation Modes per Order Quantities
## **</font></font>**

```python
##  USING SEABORN

# Font styles
font1={'family':'serif','color':'m','size':10}
font2={'family':'serif','color':'r','size':10}

transport_data = df.groupby('TRANSPORTATION_MODES')['ORDER_QUANTITIES'].sum().reset_index()


# Set plot style
sns.set(style="whitegrid")

# Create barplot
plt.figure(figsize=(6, 5))
sns.barplot(x='TRANSPORTATION_MODES', y='ORDER_QUANTITIES', data=transport_data,color="g")
plt.title("ORDER QUANTITIES BY TRANSPORTATION MODES",fontdict=font1)
plt.xlabel("Transportation Mode",fontdict=font2)
plt.ylabel("Total Order Quantity",fontdict=font2)
plt.tight_layout()
plt.show()
```

# **<font color="Green">**<font face="Times New Roman">11. Visualizing Boxplots of Defect Rates
## **</font></font>**

```python
font1={'family':'serif','color':'b','size':10}


Q1 = df['DEFECT_RATES'].quantile(0.25)
Q3 = df['DEFECT_RATES'].quantile(0.75)
IQR = Q3 - Q1

# Calculate bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['DEFECT_RATES'] < lower_bound) | (df['DEFECT_RATES'] > upper_bound)]

# View the outliers
print(outliers[['DEFECT_RATES']])

plt.figure(figsize=(5, 5))
sns.boxplot(x=df['DEFECT_RATES'], color='m')
plt.title("Boxplot of Defect Rates",fontdict=font1)
plt.xlabel("Defect Rates",fontdict=font1)

plt.show()
```
